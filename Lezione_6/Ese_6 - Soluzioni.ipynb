```python
import numpy as np

def stima_ordine(xk,iterazioni):
     #Vedi dispensa allegata per la spiegazione

      k=iterazioni-4
      p=np.log(abs(xk[k+2]-xk[k+3])/abs(xk[k+1]-xk[k+2]))/np.log(abs(xk[k+1]-xk[k+2])/abs(xk[k]-xk[k+1]));
     
      ordine=p
      return ordine
```

Scrivere la function `NewtonRaphson.py`, che implementa il metodo di Newton-Raphson

In output devono essere restituiti il vettore contenente l'approssimazione dello zero x, un vettore contenente l'errore relativo tra due iterati successivi, il numero di iterazioni

Risolvere i seguenti tre sistemi non lineari:


$$
\begin{align}

&1 = 
\begin{cases}
2x_{0} - \cos(x_{1}) = 0 \\
\sin(x_{0})+2x_{1} 0 
\end{cases} \\  
&2 = 
\begin{cases}
x_0^2+x_1^2-9=0 \\
x_0^2+2x_1-2=0
\end{cases} \\
&3 = 
\begin{cases}
4x_0^2+x_1^2=4 \\

x_0+x_1= \sin(x_0-x_1)
\end{cases} \\
\end{align}
$$


Per determinare l'iterato iniziale $x_0$ , servirsi del metodo grafico per individuare un 
iterato iniziale nell'intorno della soluzione che si vuole approssimare. Visualizzare le 
curve di livello corrisponenti a z=O delle due superfici $z = f(x_1 ,x_2 )$ e $z = h(x_1 , x_2 )$ e 
definire come iterato iniziale un vettore $x_0$ oppurtuno le cui componenti stanno in un 
intorno della soluzione (punti di intersezione tra le curve di livello delle due superfici) 
 

```python
import numpy.linalg as npl
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definizione dell'intervallo per x e y
x = np.arange(-10, 10, 0.1)
y = np.arange(-10, 10, 0.1)
X, Y = np.meshgrid(x, y)

# Definizione delle variabili simboliche
x1, x2 = sym.symbols('x1 x2')

# Scelta del sistema
scelta_sistema = int(input("Scelta sistema"))

if scelta_sistema == 1:
    F_sym = sym.Matrix([
        2 * x1 - sym.cos(x2),
        sym.sin(x1) + 2 * x2
    ])
elif scelta_sistema == 2:
    F_sym = sym.Matrix([
        x1**2 + x2**2 - 9,
        x1 + x2 - 3
    ])
elif scelta_sistema == 3:
    F_sym = sym.Matrix([
        4 * x1**2 + x2**2 - 4,
        x1 + x2 - sym.sin(x1 - x2)
    ])
elif scelta_sistema == 4:
    F_sym = sym.Matrix([
    x1**2 + x1*x2 -10,
    x2 + 3*x1*(x2**2) -57
])

# Estrazione delle funzioni
F1 = F_sym[0]
F2 = F_sym[1]

# Conversione delle funzioni simboliche in lambda function
fun1 = sym.lambdify((x1, x2), F1)
fun2 = sym.lambdify((x1, x2), F2)

# Calcolo della Jacobiana simbolicamente
JacSym = F_sym.jacobian([x1, x2])
JacNotNp = sym.lambdify((x1, x2), JacSym, 'numpy')
Jac = lambda x: np.array(JacNotNp(x[0], x[1]))

# Conversione della funzione simbolica F in lambda function
F_np = sym.lambdify((x1, x2), F_sym, 'numpy')
F = lambda x: np.array(F_np(x[0], x[1])).flatten()

# Stampa della matrice Jacobiana simbolicamente
print("La matrice Jacobiana è:")
sym.pretty_print(JacSym)

# Calcolo delle superfici Z e Z1
Z, Z1 = fun1(X, Y), fun2(X, Y)

# Plot delle superfici
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap=plt.cm.viridis)
plt.show()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z1, cmap=plt.cm.viridis)
plt.show()

```
